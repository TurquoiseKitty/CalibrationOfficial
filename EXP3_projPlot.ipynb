```python
import pandas as pd

DE_df = pd.read_csv("Experiments/EXP3/record_bin/DE_proj_benchmarks.csv").set_index("idxes")

RF_df = pd.read_csv("Experiments/EXP3/record_bin/RFKernel_proj_benchmarks.csv").set_index("idxes")

vanillaK_df = pd.read_csv("Experiments/EXP3/record_bin/vanillaKernel_proj_benchmarks.csv").set_index("idxes")
```

```python
import matplotlib.pyplot as plt

x = ["4", "10", "20", "50", "100", "150"]
MACE_DE = DE_df.loc["MACE_muSigma"][["comp_4_mu", "comp_10_mu", "comp_20_mu", "comp_50_mu", "comp_100_mu", "comp_150_mu"]]
MACE_K = vanillaK_df.loc["MACE_Loss"][["comp_4_mu", "comp_10_mu", "comp_20_mu", "comp_50_mu", "comp_100_mu"]]
MACE_RF = RF_df.loc["MACE_Loss"][["comp_4_mu", "comp_10_mu", "comp_20_mu", "comp_50_mu", "comp_100_mu", "comp_150_mu"]]

plt.plot(x, MACE_DE, label = "MACE_DE")
plt.plot(x[:-1], MACE_K, label = "MACE_K")
plt.plot(x, MACE_RF, label = "MACE_RF")
plt.legend()
plt.show()
```
